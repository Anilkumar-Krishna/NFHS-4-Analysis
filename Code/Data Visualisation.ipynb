# <font size=11> Data Visualisation </font>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= pd.read_csv('./../Data/working_data/data_rural_urban_4.csv')
```

### Indian states have  been divided in 6 zones as per the " States Reorganization Act ,1956". The district wise count and hence frequency of data collected is shown below.

```python
plt.figure(figsize=(15,15))
sns.barplot(x=data['Zone'].value_counts().index,
              y=data['Zone'].value_counts().values)
plt.xlabel('Zone')
plt.ylabel('DISTRICTS')
plt.title('DISTRICT WISE DATA FROM EACH ZONE')
plt.savefig('./../Plots/plot1.png')
plt.show()
```

### Education is often considered as key factor of general awareness . The plot below shows the state of  Male and Female Literacy in India. The mean for male literacy as per NFHS4= 85.7% while for women=68.4%

```python
sns.kdeplot(data['13. Men who are literate (%)'],shade=True,color='r')
sns.kdeplot(data['12. Women who are literate (%)'],shade=True,color='b')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Male and Female literacy in india Kde Plot System Analysis')
plt.savefig('./../Plots/plot2.png')
plt.show()
```

### Educated couples are often expected to have contraception awareness . The plot below shows how literacy in men and women across INDIA affect their practises in contraception choices . 

```python
ax = plt.gca()
```

```python
ax.set(xlim=(0,100),ylim=(0,100))
sns.lmplot(x='12. Women who are literate (%)',y='19. Any modern method (%)',data=data)
plt.xlabel('% of Female Literacy ')
plt.ylabel('Any Modern Method %')
plt.title('Sexual awareness vs % literacy in India')
plt.savefig('./../Plots/plot3.png')
sns.lmplot(x='13. Men who are literate (%)',y='19. Any modern method (%)',data=data)
plt.xlabel('% of Male Literacy ')
plt.ylabel('Any Modern Method %')
plt.title('Sexual awareness vs % literacy in India')
plt.savefig('./../Plots/plot4.png')

plt.show()
```

### Now that the above plot establishes that their indeed is a positive relation between literacy (for both men and women)  and contraceptive choice, let us dig deeper and analyse this behaviour for two  major household types: 
### 1. Urban 
### 2. Rural

```python
ax.set(xlim=(0,100),ylim=(0,100))
sns.lmplot(x='12. Women who are literate (%)',y='19. Any modern method (%)',hue='Region',data=data)
plt.xlabel('% of Female Literacy ')
plt.ylabel('Any Modern Method %')
plt.title('Sexual awareness vs % literacy in India')
plt.savefig('./../Plots/plot5.png')
sns.lmplot(x='13. Men who are literate (%)',y='19. Any modern method (%)',hue='Region',data=data)
plt.xlabel('% of Male Literacy ')
plt.ylabel('Any Modern Method %')
plt.title('Sexual awareness vs % literacy in India')
plt.savefig('./../Plots/plot6.png')
plt.show()
```

###  Here we find an interesting observation. Irrespective of gender,It is the RURAL HOUSEHOLDS that are showing a positive relationship between literacy and any mordern contr. method while the URBAN HOUSEHOLDS are showing a decline.
