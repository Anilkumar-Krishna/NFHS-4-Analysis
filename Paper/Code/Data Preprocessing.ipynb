## Merging the data from the 30 files into one single file, so as to facilitate analysis on an all India level 

## DATA PREPROCESSING: Feature Construction (State)

```python
import os
import io
import pandas as pd
flag=0
for root, dirs, files in os.walk("./../Data/original_data"):
    for filename in files:
          name="./../Data/original_data" + "/"  +filename
          df=pd.read_csv(name)
          #print(df)
          df=df.transpose()
          df.columns = df.iloc[0]
          df=df.drop(df.index[0])
          statename=[]
          df=df.dropna(how='all')
          n=df.shape[0]

          name = filename.strip('.csv')
          for i in range(n):
            statename.append(name)

          df['State']=statename
          if(flag==0):
            dataset=df
            flag=1
          else:
            dataset=dataset.append(df)

```

## DATA PREPROCESSING: Feature Contruction: Region

```python
region=[]
district=[]

n=dataset.shape[0]
for i in range(n):
    
    name=dataset.iloc[i]['District']
    
    if(name.find('Rural')!=-1):
        
        name=name.replace('- Rural', '')
        name=name.replace('-Rural', '')
        
        name=name.strip()
        district.append(name)
        region.append('Rural')
        
    elif(name.find('Urban')!=-1):
        name=name.replace('- Urban', '')
        name=name.replace('-Urban', '')
        name=name.strip()
        district.append(name)
        region.append('Urban')
        
    elif(name.find('Total')!=-1):
        name=name.replace('- Total', '')
        name=name.replace('-Total', '')
        name=name.strip()
        district.append(name)
        region.append('Total')
       
print(len(region))
print(len(district))
print(dataset.shape[0])

```

```python
dataset['Region']=region
dataset['District']=district
dataset
```

## DATA PREPROCESSING: Feature Construction : Zone

```python
zone={ 'Southern Zone' : ['AP', 'TN' , 'KL', 'KA', 'TG'] , 'Western Zone' : ['GA', 'MH', 'GJ'] , 'Eastern Zone' : ['OR','WB', 'BR', 'JH', 'SK'] , 'North Eastern Zone' : ['AR','NL','MN','MZ','ML','AS','TR'], 'Central Zone' : ['MP' , 'CT' , 'UP' , 'UT'], 'Northern Zone' : ['RJ', 'DL' , 'PB' , 'HR' , 'HP' , 'JK']}
```

```python
zone_list=[]
n=dataset.shape[0]
for i in range(n):
    name=dataset.iloc[i]['State']
    for z in zone:
        if name in zone[z]:
            zone_list.append(z)
print(len(zone_list))
print(n)
    
```

```python
dataset['Zone']=zone_list
dataset
```

## DATA PREPROCESSING: Unsupervised Discretisation : Equal Interval Binning

```python
dataset4=dataset3.copy()
to_ignore = ['District', 'State', 'Region', 'Zone']
for col in dataset4.columns:
    if col not in to_ignore:
        dataset4[col]=pd.cut(dataset4[col],3,labels=['low', 'medium', 'high'])
    

```

## DATA PREPROCESSING: Unsupervised Discretisation : Equal Frequency Binning
